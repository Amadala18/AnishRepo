With this in mind, we can start to become more specific in our research.  Instead of gathering data about the pandemic from its beginning and its end, we should also gather data from during the pandemic to get a more accurate trend line with teleworking rates:


**How did COVID-19 impact the rate of people working remotely, and during which periods the rate fluctuate the most?**

In order to start answering this question, we gather data from around the beginning of the pandemic, May 2020, to the end, September 2022; this is **29 months** of data.  To speed up the parsing process, we can create a for loop to iterate through the files, and create a **DataFrame object** with all of the extracted values, including the *percent of people teleworking*, and then *time period*.  We can sort these values by *time period* so we can start plotting a line graph in chronological order.

```python
import os

dir_path = 'Q1 CSV'
files = os.listdir(dir_path)

dataPeriods = []
dataValues = []
dataValuesDF = []

for x in range(len(files)):
    fileName = files[x]
    dataName = (fileName.split('-',2)[2]).split('.')[0]
    dataNameYear = dataName.split('-')[0].split('20', 1)[1]
    dataNameMonth = dataName.split('-')[1]
    dataNameFinal = str(dataNameYear) + '/' + str(dataNameMonth)
#     print(dataNameFinal)
    dataPeriods.append(dataNameFinal)
    csvName = dir_path + '/' + fileName
    data = pandas.read_csv(csvName)
    data = data.rename(columns={data.columns[0]: "Characteristics", data.columns[1]: "Total Employed in Time Period", data.columns[2]: "Total persons teleworking due to pandemic", data.columns[3]: "Percent of Total Employed", data.columns[4]: "Percent distribution of total employed", data.columns[5]: "Percent distribution of persons teleworking due to pandemic"})
    dataValues = data['Percent of Total Employed'].dropna().reset_index()
    datatotalEmp = dataValues['Percent of Total Employed'][1]
    dataValuesDF.append(float(datatotalEmp))

graph2 = pandas.DataFrame({'Time_Period': dataPeriods,
        'Percent_Teleworking': dataValuesDF})

graph2 = graph2.sort_values(by= 'Time_Period').reset_index(drop = True).set_index('Time_Period')

graph2
```

Above, we see the final DataFrame object, sorting chronologically by time period during the COVID-19 pandemic.  In order to conduct analysis on these values, we can plot these values in a line graph.

Given our DataFrame, we can start plotting our values on a line graph.  In the code below, we parse through the DataFrame for obvious peaks and troughs in our data, which represent high and low percentages of teleworking Americans.  For our visualization, the peaks are represented with **<span style="color:green">green</span>** points, and the troughs are represented with **<span style="color:red">red</span>** points.

```python
from scipy.signal import argrelextrema
import math

ax = plt.figure(figsize=(12.5,6))

 
# using rc function

n = 2  # number of points to be checked before and after

# Find local peaks
graph2['min'] = graph2.iloc[argrelextrema(graph2.Percent_Teleworking.values, np.less_equal,
                    order=n)[0]]['Percent_Teleworking']
graph2['max'] = graph2.iloc[argrelextrema(graph2.Percent_Teleworking.values, np.greater_equal,
                    order=n)[0]]['Percent_Teleworking']

minsA = np.where(graph2['min'] > 0)[0]
maxesA = np.where(graph2['max'] > 0)[0]

default_x_ticks = np.concatenate((minsA, maxesA))


plt.scatter(graph2.index, graph2['min'], c='r', label='trough')
plt.scatter(graph2.index, graph2['max'], c='g', label='peak')
plt.plot(graph2.index, graph2['Percent_Teleworking'])

plt.xticks(default_x_ticks, fontsize = 7)
# setting title name
plt.title("Rate of People Teleworking during COVID-19 Pandemic")
 
# setting x axis label
plt.xlabel("Time Period", fontsize = 20)
 
# setting y axis label
plt.ylabel("Percentage", fontsize = 20)

for x in default_x_ticks:
    plt.text(x+0.1, graph2['Percent_Teleworking'][x] - 0.5, str(graph2['Percent_Teleworking'][x]), fontsize=17, weight = 'bold')
plt.legend(loc='upper right')


y = [(graph2['Percent_Teleworking'][0]), graph2['Percent_Teleworking'][28]]
x = [0, 28]
# plt.plot(x, y, c = 'red', linestyle = '-.' )

# Calculate the Trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

# Display the Trendline
plt.plot(x, p(x), c = 'r', linestyle='--',)

plt.show()


# graph.plot(xticks = [0, 28], fontsize = 9, style = 'bo-', ylabel = 'Percent')
```

With the additional data analyzed, we can now see dips and highs within the teleworking rates throughout the entirety of the pandemic.  The first rate recorded was in *May 2020*, with a high percentage of **35.4%**.  From there, we see the rate decrease in the following months, reaching a low percentage of **21.2%** in *October 2020*.  Without external research and from personal experience, we can hypothesize that the general constant decline in percentage between these two months can be attributed to the introduction of masks, which led to the lower percentage of COVID-19 reports.  Due to the lower percentages of reports, more and more people may have returned to work, which decreases the teleworking rate.

However, we see a spike in the teleworking rate, **23.7%** in *December 2020*, around 2 months after the first trough in *October 2020*.  Considering the spike 2 months later, we can start to hypothesize that, due to the return of the working population back to commuting in-office, the number of COVID-19 reports starting rising again, which led to some people returning to working from home. Additionally, these are stats for the wintertime, in which sickness rates already increase due to the cold weather.

After this spike, we can see that the teleworking decreases gradually until *December 2021*.  After this, in *January 2022*, the rate suddenly increases by around *4.4%* to **15.4%**.  Just like the spike between *October 2020* and *December 2020*, we can hypothesize that the number of COVID-19 cases rose due to winter weather conditions.

From this point on, the rate decreased to **5.2%** in *September 2022*, which was the final time period that we have data for.

As we can see from the **<span style="color:red">red</span>** trend line, the overall rate decreases; but from the line plotted from the 29 csv files, we can look at the trends more accurately, and see the dips.  From the line, we are able to tell the periods of time in which the rates fluctuated the most.

# Research Question 4 
How did COVID-19 impact the rate of unemployment in the United States?

```python
import pandas
# Data is from Bureau of Labor Statistics (BLS) https://www.bls.gov/data/
umemployment_data = {'Year': ['2020']*12 + ['2021']*12,
        'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']*2,
        'Unemployment Rate': [3.6, 3.5, 4.4, 14.8, 13.3, 11.1, 10.2, 8.4, 7.9, 6.9, 6.7, 6.7,
                              6.3, 6.2, 6.0, 14.8, 13.0, 11.1, 10.2, 8.4, 7.9, 6.9, 6.7, 6.7]}

unemployment_table = pandas.DataFrame(umemployment_data)
print(unemployment_table)
```

The above table shows the unemployment rates for each month throughout the months of 2020 and 2021. These are the prime months, specifically early 2020 and 2021 when unemployment had a big hit as a result of COVID-19. From the table, we can see that the starting from April of 2020, the unemployment rate almost went up by a rate of 5 times the previos month and remained to be at twice the rate for the rest of the months. There was another spike in April of 2021 and although it went down, we will never know when the rate of unemployment will reach pre-COVID-19 rates.

```python
#Predicting next 24 months
from sklearn.linear_model import LinearRegression

months = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
unemployment_table['Month'] = unemployment_table['Month'].map(months)
x = unemployment_table[['Year', 'Month']]
y = unemployment_table['Unemployment Rate']
model = LinearRegression()
model.fit(x, y)
predictions = pandas.DataFrame({'Year': ['2022']*12 + ['2023']*12,
                               'Month': list(range(1, 13))*2})
x_pred = predictions[['Year', 'Month']]
y_pred = model.predict(x_pred)
predictions['Unemployment Rate'] = y_pred
print(predictions)
```

```python
#Graphing the data
import matplotlib.pyplot

combined_data = pandas.concat([unemployment_table, predictions], ignore_index = True)
matplotlib.pyplot.plot(combined_data['Year'] + '-' + combined_data['Month'].astype(str), combined_data['Unemployment Rate'])
matplotlib.pyplot.xticks(rotation = 45)
matplotlib.pyplot.xlabel('Year-Month')
matplotlib.pyplot.ylabel('Unemployment Rate')
matplotlib.pyplot.title('Unemployment Rates from 2020 to 2023')
matplotlib.pyplot.show()
```

After seeing the graph, we can see that unemployment spiked greatly during 2020 and 2021 but even though it has not reached to the level it was before COVID-19, it has leveled out over the rest of 2023 but it is still high at a rate of about 9.3%. Will the rate ever get to below 4% to how it was before COVID times? This question is hard to answer as the rate is relatviely constant but at almost 10% and with predictions of a recession coming and mass layoffs, many predict that this rate will only get higher.

```python
#Prediting when the unemployment rate will reach 5%
pre_covid_unemployment_rate = 5
intercept = model.intercept_
coef = model.coef_
slope = coef[0] * 12 + coef[1]
months_until_steady_rate = (pre_covid_unemployment_rate - intercept) / slope
print('Unemployment rates will steadily reach 5% in', int(months_until_steady_rate), 'months.')
```
